```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('./titanic.csv')
dataset.head(3)
```

```python
##Calculating Mean 
dataset['Age'].mean()
mn=np.mean(dataset['Age'])
```

```python
# Create a histogram using seaborn's histplot function
# x='Age' specifies the column to plot on the x-axis
# data=dataset specifies the DataFrame containing the data
# bins=[i for i in range (0,81,10)] sets the bin edges explicitly
# The bins range from 0 to 80, with a width of 10 for each bin (0-10, 10-20, ..., 70-80)
sns.histplot(x='Age',data=dataset,bins=[i for i in range (0,81,10)])

# The following line attempts to plot a straight line on the graph
# [mn for i in range (0,300)] creates a list of 300 repetitions of the variable 'mn' for the x-coordinates
# [i for i in range (0,300)] creates a list of numbers from 0 to 299 for the y-coordinates
# This likely requires 'mn' to be defined elsewhere and might not plot a meaningful line in this context
# The 'c='yellow'' argument sets the line color to yellow
plt.plot([mn for i in range (0,300)],[i for i in range (0,300)],c='yellow')

# Display the generated plot and any lines drawn on it
plt.show()
```
